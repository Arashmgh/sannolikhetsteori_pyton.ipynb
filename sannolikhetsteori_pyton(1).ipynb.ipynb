```python
## Ilustrate the Law of Large Numbers (LLN)  [Called "De relative frekvensernas stabilitet" in the book p.14]
#1. Simulate 10000 dice throws. 
#2. Plot the relative frequency of the number of times three dots occur for 1, 2, ..., 10000 throws.
#3. To what number does the relative frequency converge to? What is the probability of getting three dots when throwing a dice? Do these two numbers coincide?
#4. Explain what you observe. 

#Make sure to create a good pedagogical plot, i.e. include title, x-label and y-label. 
# Repeating the Law of Large Numbers (LLN) simulation with detailed comments and pedagogical plotting

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Setting up the parameters
num_throws = 10000  # Number of dice throws
target_outcome = 3  # Outcome to track (three dots)

# Simulate dice throws, each throw randomly gives an integer from 1 to 6
throws = np.random.randint(1, 7, num_throws)

# Calculate the cumulative count of throws where the outcome was 3 dots, and compute relative frequency
relative_frequency = np.cumsum(throws == target_outcome) / np.arange(1, num_throws + 1)

# Calculate the theoretical probability of getting three dots (1/6 for a fair die)
expected_probability = 1 / 6

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(relative_frequency, color="blue", label="Relative Frequency of Three Dots")
plt.axhline(expected_probability, color='red', linestyle='--', label="Expected Probability (1/6)")

# Adding labels, title, legend, and grid for better understanding
plt.title("Law of Large Numbers: Convergence of Relative Frequency of Rolling Three Dots")
plt.xlabel("Number of Throws")
plt.ylabel("Relative Frequency of Three Dots")
plt.legend()
plt.grid(True)
plt.show()

# Summary observations based on this simulation:
relative_frequency[-1], expected_probability


```

```python
# We know that the binomial coefficients can be used to calculate the number of ways to choose k elements from n without replacement when the order does not matter. We have the formula: $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ .

# Now calculate $\binom{6}{2}$ with a built in function and then create your own function (you can use the built in factorial function) and verify that you get the same results. 
import math
from scipy.special import comb  # For using built-in function to calculate binomial coefficient

# Using built-in function to calculate C(6,2)
built_in_result = comb(6, 2, exact=True)

# Defining custom function to calculate binomial coefficient
def binomial_coefficient(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

# Calculating C(6,2) using the custom function
custom_result = binomial_coefficient(6, 2)

built_in_result, custom_result
```

```python
# Simulate 10000 samples from bin(n= 100, p = 0.35) and plot a histogram. Then simulate 10000 samples from bin(n= 100, p = 0.9) and plot a histogram. In the plots, what is the "middle" of the plot? Which plot has a "wider" distribution?
# What is the theoretical/true Expected Value and Variance of the two distributions? (see p.89 in the book for formulas.)
# Calculate the mean and variance of the two samples that you simulated, use the functions np.mean() and np.var(). Explain your observations .
# Re-importing necessary libraries after reset
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distributions
n1, p1 = 100, 0.35  # For the first distribution
n2, p2 = 100, 0.9   # For the second distribution
num_samples = 10000  # Number of samples

# Simulate samples for both distributions
samples1 = binom.rvs(n=n1, p=p1, size=num_samples)
samples2 = binom.rvs(n=n2, p=p2, size=num_samples)

# Plot histograms for both distributions
plt.figure(figsize=(14, 6))

# Histogram for the first distribution (n=100, p=0.35)
plt.subplot(1, 2, 1)
plt.hist(samples1, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Binomial Distribution (n=100, p=0.35)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)

# Histogram for the second distribution (n=100, p=0.9)
plt.subplot(1, 2, 2)
plt.hist(samples2, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title("Histogram of Binomial Distribution (n=100, p=0.9)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculating theoretical expected value and variance for both distributions
expected_value1, variance1 = n1 * p1, n1 * p1 * (1 - p1)
expected_value2, variance2 = n2 * p2, n2 * p2 * (1 - p2)

# Calculating sample mean and variance for both distributions
sample_mean1, sample_variance1 = np.mean(samples1), np.var(samples1)
sample_mean2, sample_variance2 = np.mean(samples2), np.var(samples2)

# Outputting the results
((expected_value1, variance1, sample_mean1, sample_variance1),
 (expected_value2, variance2, sample_mean2, sample_variance2))

```

```python
# Solve exercise 307 in Python. 
from scipy.stats import binom

# Parameters for the binomial distribution
n, p = 12, 0.4

# (a) Pr(X <= 6)
prob_a = binom.cdf(6, n, p)

# (b) Pr(X < 6)
prob_b = binom.cdf(5, n, p)

# (c) Pr(X = 6)
prob_c = binom.pmf(6, n, p)

# (d) Pr(X > 6)
prob_d = 1 - prob_a

prob_a, prob_b, prob_c, prob_d

```
